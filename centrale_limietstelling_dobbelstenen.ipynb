```python
#Imports en helperfuncties
import random
from scipy import stats as ss
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.pylab as pylab

%matplotlib inline
```

```python
hu_rood = "#d7000d"
hu_blauw = "#00a0d2"
mpl.style.use(['seaborn-notebook', 'seaborn-darkgrid'])

params = {'xtick.labelsize': 'xx-large', 'ytick.labelsize': 'xx-large', 'figure.titlesize': 'xx-large'}
pylab.rcParams.update(params)

fsize, fdpi = (10,6), 150
```

```python
aantalworpen = [1, 10, 100, 1000]
worpen = [np.random.randint(1, 7, size=(1000, i)) for i in aantalworpen]
sommen = [worp.mean(axis=1) for worp in worpen]

plt.figure(figsize=fsize, dpi=fdpi)
for i in range(0, 4):
    h = aantalworpen[i]
    ax = plt.subplot(2, 2, i+1)
    plt.hist(sommen[i], bins=max(6, min(25, int(np.sqrt(h)))), alpha=0.5)
    plt.title("{} worp(en), gemiddelde = {:.2f}".format(h, sommen[i].mean()))
    
plt.tight_layout()
```
