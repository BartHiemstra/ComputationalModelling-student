```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
# vul de punten RGB-waardes
RGB = [[247,247,247],
       [255,247,255],
       [234,234,234],
       [221,221,221],
       [247,234,247],
       [234,221,234],
       [255,255,255],
       [234,234,247],
       [234,221,221],
       [78,78,65],
       [65,65,52],
       [221,208,208],
       [78,78,52],
       [234,182,156],
       [247,247,255],
       [221,182,143],
       [104,78,52],
       [78,52,39],
       [104,78,65]]
```

```python
# Ter illustratie de kleuren
fig, ax = plt.subplots()

i = 0
for color in RGB:
    circle = plt.Circle((i*0.2, 0.5), 0.1, color=[c / 256 for c in color], clip_on=False)
    ax.add_artist(circle)
    i += 1
    
plt.axis('off')
plt.show()
```
